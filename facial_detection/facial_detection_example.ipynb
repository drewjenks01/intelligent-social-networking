```python
from facial_detection import FacialDetection
from PIL import Image
```

```python
facial_detection = FacialDetection(visualize=True)
```

```python
example_img = Image.open('example_facial_img.png').convert('RGB')
example_img.resize((400, 400))
```

```python
example_img_2 = Image.open('example_facial_img2.png').convert('RGB')
example_img_2.resize((400,400))
```

```python
import matplotlib.pyplot as plt
detected_face = facial_detection.detect_face(example_img)

plt.imshow(detected_face[0].permute(1, 2, 0).int().numpy())
plt.axis('off')
plt.show()


detected_face_2 = facial_detection.detect_face(example_img_2)

plt.imshow(detected_face_2[0].permute(1, 2, 0).int().numpy())
plt.axis('off')
plt.show()
```

```python
# need to make a new object without visualize=True in order to get embeddings
facial_detection = FacialDetection(visualize=False)
detected_face = facial_detection.detect_face(example_img)
embedding = facial_detection.get_facial_embeddings(detected_face)
```

```python
embedding.shape
```

```python
detected_face_2 = facial_detection.detect_face(example_img_2)
embedding_2 = facial_detection.get_facial_embeddings(detected_face_2)
```

```python
embedding_dis = facial_detection.get_embedding_distance(embedding, embedding_2)
embedding_dis
```

```python
example_img_stephen = Image.open('example_facial_stephen.png').convert('RGB')
example_img_stephen.resize((400, 400))
```

```python
detected_face_stephen = facial_detection.detect_face(example_img_stephen)
embedding_stephen = facial_detection.get_facial_embeddings(detected_face_stephen)
distance = facial_detection.get_embedding_distance(embedding, embedding_stephen)
```

```python
distance
```

```python
# lower distance = more similar
print(f'Distance between 2 Drew images: {embedding_dis}')
print(f'Distance between Drew and Stephen images: {distance}')
```
